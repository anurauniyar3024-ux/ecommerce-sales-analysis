## Handling Missing Values - Orders

The missing values in delivery-related columns were analyzed before cleaning.

Most missing timestamps belong to orders with statuses such as **canceled**, **processing**, **shipped**, and **unavailable**, where missing delivery information is expected as part of the business process.

Since these missing values represent real operational scenarios rather than data quality issues, no imputation or deletion was performed.

```python
products.isnull().sum()
```

```python
products["product_weight_g"] = products["product_weight_g"].fillna(products["product_weight_g"].median())

products["product_length_cm"] = products["product_length_cm"].fillna(products["product_length_cm"].median())

products["product_height_cm"] = products["product_height_cm"].fillna(products["product_height_cm"].median())

products["product_width_cm"] = products["product_width_cm"].fillna(products["product_width_cm"].median())

# Verify
products.isnull().sum()
```

```python
products.isnull().sum()
```

```python
datasets = {
    "Customers": customers,
    "Orders": orders,
    "Order Items": order_items,
    "Payments": payments,
    "Reviews": reviews,
    "Products": products,
    "Sellers": sellers,
    "Geolocation": geolocation,
    "Category Translation": category
}

for name, df in datasets.items():
    print(f"{name}: {df.duplicated().sum()} duplicate rows")
```

## Duplicate Records

A duplicate check was performed on all datasets.

- Customers, Orders, Order Items, Payments, Reviews, Products, Sellers, and Category Translation contained **no duplicate records**.
- The Geolocation dataset contains **261,831 duplicate rows**.

The geolocation duplicates were not removed during this stage because repeated geographic records can be valid and may be useful for future location-based analysis. Since this project focuses on sales analytics, the dataset is preserved in its original form.

```python
customers.to_csv("../data/cleaned/customers_cleaned.csv", index=False)
orders.to_csv("../data/cleaned/orders_cleaned.csv", index=False)
order_items.to_csv("../data/cleaned/order_items_cleaned.csv", index=False)
payments.to_csv("../data/cleaned/payments_cleaned.csv", index=False)
reviews.to_csv("../data/cleaned/reviews_cleaned.csv", index=False)
products.to_csv("../data/cleaned/products_cleaned.csv", index=False)
sellers.to_csv("../data/cleaned/sellers_cleaned.csv", index=False)
geolocation.to_csv("../data/cleaned/geolocation_cleaned.csv", index=False)
category.to_csv("../data/cleaned/category_translation_cleaned.csv", index=False)

print("✅ All cleaned datasets have been saved successfully!")
```
